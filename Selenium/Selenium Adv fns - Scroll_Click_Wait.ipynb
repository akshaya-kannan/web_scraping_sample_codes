### Create a web driver object by pointing it to the location of the chromedriver program

```python
driver = webdriver.Chrome(r'C:\Users\anirudh.rao\Desktop\chromedriver.exe')
```

### Get the URL using get method

```python
urls="https://www.chicagolandchamber.org/Membership/MemberDirectory.aspx?desc=&mbg=&submit=Search+Members"
driver.get(urls)
```

```python
import time
for i in range(1,400):
    
    while True:
        driver.execute_script("window.scrollTo(0, 2)")
        time.sleep(1)
    
        rows=WebDriverWait(driver, 10).until(lambda x: x.find_elements_by_xpath("//div[@class='featured-member clear']//div[@class='details']"))
        print(rows)
        for row in rows:
            try:
                name=row.find_element_by_xpath(".//h3").text
            except:
                name=''
            print(name)    
            try:
                web=WebDriverWait(row, 10).until(lambda x: x.find_element_by_xpath(".//a")).get_attribute("href")
            except:
                web=''    
            print(web)    
            if (name !=None):
                resultDict = {"chamber_name" : "Brazil Chicagoland Chamber of Commerce Singapore",
                        "country" : 'USA',
                        "region" : 'America',
                        "base_url" : "https://www.chicagolandchamber.org",
                        "source_url" : urls,
                        "company_website" : web,
                        "company_contact" : None,
                        "company_name" : name,
                        "company_address" : None,
                          "company_category":None

                           }
                print(resultDict)
                df = pd.DataFrame([resultDict])
                df.to_csv('brazil_chicago_1.csv', mode='a', header=False)
        WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//li[@class = 'next']/a"))).click()
        
        
```
